# Foundations of Statistical Inference & Hypothesis Testing

This section checks the statistical foundation behind baseline monitoring. The main idea is to understand how sample estimates behave, why covariance estimation needs Bessel's correction, how hypothesis testing controls false alarms, and how first-moment homogeneity can be checked numerically using Wilks' Lambda.

## Part A: Theoretical Fundamentals  
### 1. Bias of the Maximum Likelihood Covariance Estimator

Let the observations be

\[
\mathbf{X}_1,\mathbf{X}_2,\ldots,\mathbf{X}_n \in \mathbb{R}^m
\]

with population mean

\[
\boldsymbol{\mu}=E[\mathbf{X}_i]
\]

and population covariance matrix

\[
\boldsymbol{\Sigma}=Var(\mathbf{X}_i).
\]

The sample mean is

\[
\widehat{\boldsymbol{\mu}}_n=\frac{1}{n}\sum_{i=1}^{n}\mathbf{X}_i.
\]

The raw maximum likelihood estimator of the covariance matrix is

\[
\widehat{\boldsymbol{\Sigma}}_{MLE}
=
\frac{1}{n}\sum_{i=1}^{n}
(\mathbf{X}_i-\widehat{\boldsymbol{\mu}}_n)
(\mathbf{X}_i-\widehat{\boldsymbol{\mu}}_n)^T.
\]

To examine its expectation, use the decomposition

\[
\mathbf{X}_i-\widehat{\boldsymbol{\mu}}_n
=
(\mathbf{X}_i-\boldsymbol{\mu})
-
(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu}).
\]

A standard identity for the scatter matrix gives

\[
\sum_{i=1}^{n}
(\mathbf{X}_i-\widehat{\boldsymbol{\mu}}_n)
(\mathbf{X}_i-\widehat{\boldsymbol{\mu}}_n)^T
=
\sum_{i=1}^{n}
(\mathbf{X}_i-\boldsymbol{\mu})
(\mathbf{X}_i-\boldsymbol{\mu})^T
-
n(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})
(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})^T.
\]

Taking expectations,

\[
E\left[
\sum_{i=1}^{n}
(\mathbf{X}_i-\boldsymbol{\mu})
(\mathbf{X}_i-\boldsymbol{\mu})^T
\right]
=
n\boldsymbol{\Sigma}.
\]

Also,

\[
Var(\widehat{\boldsymbol{\mu}}_n)=\frac{1}{n}\boldsymbol{\Sigma}.
\]

Therefore,

\[
E\left[
n(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})
(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})^T
\right]
=
n \cdot \frac{1}{n}\boldsymbol{\Sigma}
=
\boldsymbol{\Sigma}.
\]

Thus,

\[
E\left[
\sum_{i=1}^{n}
(\mathbf{X}_i-\widehat{\boldsymbol{\mu}}_n)
(\mathbf{X}_i-\widehat{\boldsymbol{\mu}}_n)^T
\right]
=
n\boldsymbol{\Sigma}-\boldsymbol{\Sigma}
=
(n-1)\boldsymbol{\Sigma}.
\]

Since the MLE divides this scatter matrix by \(n\),

\[
E[\widehat{\boldsymbol{\Sigma}}_{MLE}]
=
\frac{1}{n}(n-1)\boldsymbol{\Sigma}
=
\frac{n-1}{n}\boldsymbol{\Sigma}.
\]

Therefore, the raw MLE covariance estimator is biased downward in finite samples.

To remove this bias, the scatter matrix must be divided by \(n-1\) instead of \(n\). This gives

\[
\mathbf{S}
=
\frac{1}{n-1}\sum_{i=1}^{n}
(\mathbf{X}_i-\widehat{\boldsymbol{\mu}}_n)
(\mathbf{X}_i-\widehat{\boldsymbol{\mu}}_n)^T.
\]

Then,

\[
E[\mathbf{S}]
=
\frac{1}{n-1}(n-1)\boldsymbol{\Sigma}
=
\boldsymbol{\Sigma}.
\]

Hence, Bessel's correction produces the unbiased empirical covariance estimator.

### 2. Type I Error, Type II Error, Power, and the Healthy Operation Ellipsoid

Under the null hypothesis \(H_0\), the structural asset is assumed to remain healthy. A test statistic is monitored to decide whether the system still behaves like the baseline condition or whether it has shifted away from it.

A Type I Error occurs when the system is actually healthy, but the test rejects \(H_0\). In structural health monitoring, this means the diagnostic system raises a false alarm even though the asset has not changed. The probability of this error is denoted by

\[
\alpha.
\]

A Type II Error occurs when the system is actually damaged or shifted, but the test fails to reject \(H_0\). In this case, the monitoring system misses a real abnormal condition. The probability of this error is

\[
\beta.
\]

The statistical power of the test is

\[
1-\beta.
\]

Power is the probability that the diagnostic layer correctly detects a real shift or damage condition.

If an engineer sets an extremely small significance threshold such as

\[
\alpha=0.0001,
\]

the test becomes very conservative. This reduces the probability of false alarms, but it also makes the rejection region much smaller. As a result, the system requires stronger evidence before declaring damage.

Mathematically, lowering \(\alpha\) increases the critical threshold of the test statistic. For a multivariate monitoring statistic, the healthy region can be represented by a confidence ellipsoid of the form

\[
(\mathbf{x}-\boldsymbol{\mu})^T
\boldsymbol{\Sigma}^{-1}
(\mathbf{x}-\boldsymbol{\mu})
\leq c_{\alpha}.
\]

When \(\alpha\) is made smaller, \(c_{\alpha}\) becomes larger. This expands the healthy operation confidence ellipsoid. More observations are then classified as healthy.

The consequence is that false alarms decrease, but the probability of missing smaller real shifts increases. Therefore, \(\beta\) increases and the power \(1-\beta\) decreases. In practice, the plant becomes safer from unnecessary shutdowns, but less sensitive to early-stage damage.

## Part B: Theoretical Extension  
### Asymptotic Distributions and Slutsky's Theorem

The multivariate central limit theorem states that, for independent and identically distributed observations with finite mean and covariance,

\[
\sqrt{n}(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})
\xrightarrow{d}
\mathcal{N}(\mathbf{0},\boldsymbol{\Sigma}).
\]

This result uses the true population covariance matrix \(\boldsymbol{\Sigma}\). In practice, however, \(\boldsymbol{\Sigma}\) is unknown and must be estimated from the sample.

Slutsky's Theorem states that if

\[
A_n \xrightarrow{d} A
\]

and

\[
B_n \xrightarrow{p} b,
\]

where \(b\) is a constant, then

\[
A_n+B_n \xrightarrow{d} A+b,
\]

\[
A_nB_n \xrightarrow{d} Ab,
\]

and, if \(b\neq 0\),

\[
\frac{A_n}{B_n}\xrightarrow{d}\frac{A}{b}.
\]

In the multivariate case, the same idea applies to consistent matrix estimators. The unbiased sample covariance matrix

\[
\mathbf{S}
=
\frac{1}{n-1}\sum_{i=1}^{n}
(\mathbf{X}_i-\widehat{\boldsymbol{\mu}}_n)
(\mathbf{X}_i-\widehat{\boldsymbol{\mu}}_n)^T
\]

is a consistent estimator of the population covariance matrix:

\[
\mathbf{S}\xrightarrow{p}\boldsymbol{\Sigma}.
\]

Since the sample covariance converges in probability to the true covariance, Slutsky's Theorem justifies replacing the unknown \(\boldsymbol{\Sigma}\) with \(\mathbf{S}\) asymptotically.

Therefore, for large \(n\),

\[
\widehat{\boldsymbol{\mu}}_n
\approx
\mathcal{N}
\left(
\boldsymbol{\mu},
\frac{1}{n}\boldsymbol{\Sigma}
\right).
\]

Since \(\boldsymbol{\Sigma}\) is unknown but \(\mathbf{S}\) consistently estimates it, the practical modelling distribution becomes

\[
\widehat{\boldsymbol{\mu}}_n
\approx
\mathcal{N}
\left(
\boldsymbol{\mu},
\frac{1}{n}\mathbf{S}
\right).
\]

This is the operational form used in real monitoring systems because it replaces the unobservable population covariance with the empirical covariance estimated from baseline data.

## Part C: Numerical Verification

The first numerical task checks whether the joint mean vector remains constant over time. The dataset is divided into five consecutive blocks of 1000 observations each. A MANOVA-style test is then applied using Wilks' Lambda.

The null hypothesis is:

\[
H_0: \text{the joint mean vector is constant across all time blocks.}
\]

The alternative hypothesis is:

\[
H_1: \text{at least one time block has a shifted joint mean vector.}
\]

The test statistic is based on

\[
\Lambda = \frac{|\mathbf{W}|}{|\mathbf{B}+\mathbf{W}|},
\]

where \(\mathbf{W}\) is the within-block scatter matrix and \(\mathbf{B}\) is the between-block scatter matrix.

Bartlett's chi-square approximation is

\[
\chi^2_{calc}
=
-\left(n-1-\frac{m+g}{2}\right)\ln(\Lambda).
\]

The degrees of freedom are

\[
df=m(g-1).
\]

```python
# ============================================================
# Foundations of Statistical Inference & Hypothesis Testing
# Part C: Numerical Verification
# MANOVA first-moment homogeneity test using Wilks' Lambda
# ============================================================

import numpy as np
import pandas as pd
import scipy.stats as stats

from IPython.display import Markdown, display

# --- DO NOT MODIFY: Synthetic Strain Sensor Data Generator ---
np.random.seed(42)

n_samples = 5000
n_features = 3

base_data = np.random.multivariate_normal(
    mean=[0.5, -0.2, 1.1],
    cov=[
        [0.09, 0.02, 0.01],
        [0.02, 0.06, 0.03],
        [0.01, 0.03, 0.05]
    ],
    size=n_samples
)

# Injecting a subtle, hidden structural drift in the final 1,000 snapshots
base_data[4000:, 0] += 0.015  # Sensor 1 drift
base_data[4000:, 2] -= 0.010  # Sensor 3 drift

df_strain = pd.DataFrame(
    base_data,
    columns=["Strain_Ch1", "Strain_Ch2", "Strain_Ch3"]
)


def verify_first_moment_homogeneity(df: pd.DataFrame, g_chunks: int = 5, alpha: float = 0.05):
    """
    Partitions the dataset into g consecutive non-overlapping chunks and evaluates
    first-moment homogeneity using Wilks' Lambda and Bartlett's Chi-Square approximation.
    """

    X = df.select_dtypes(include=[np.number]).to_numpy(dtype=float)
    n, m = X.shape

    if n % g_chunks != 0:
        raise ValueError("The number of samples must be divisible by g_chunks.")

    chunk_size = n // g_chunks

    blocks = [
        X[i * chunk_size:(i + 1) * chunk_size, :]
        for i in range(g_chunks)
    ]

    global_mean = np.mean(X, axis=0)

    W = np.zeros((m, m))
    B = np.zeros((m, m))

    chunk_means = []

    for block in blocks:
        chunk_mean = np.mean(block, axis=0)
        chunk_means.append(chunk_mean)

        centered_block = block - chunk_mean
        W += centered_block.T @ centered_block

        mean_difference = (chunk_mean - global_mean).reshape(-1, 1)
        B += block.shape[0] * (mean_difference @ mean_difference.T)

    T = B + W

    sign_W, logdet_W = np.linalg.slogdet(W)
    sign_T, logdet_T = np.linalg.slogdet(T)

    wilks_lambda = np.exp(logdet_W - logdet_T)

    chi_square_calc = -(n - 1 - ((m + g_chunks) / 2)) * np.log(wilks_lambda)

    df_chi_square = m * (g_chunks - 1)

    p_value = stats.chi2.sf(chi_square_calc, df=df_chi_square)

    if p_value < alpha:
        conclusion = "Reject H0: the baseline center has shifted over time."
    else:
        conclusion = "Fail to reject H0: the first moment is homogeneous."

    results = {
        "n_samples": n,
        "n_features": m,
        "g_chunks": g_chunks,
        "chunk_size": chunk_size,
        "Wilks_Lambda": wilks_lambda,
        "Bartlett_Chi_Square": chi_square_calc,
        "Degrees_of_Freedom": df_chi_square,
        "p_value": p_value,
        "alpha": alpha,
        "Conclusion": conclusion
    }

    chunk_means_df = pd.DataFrame(
        chunk_means,
        columns=df.select_dtypes(include=[np.number]).columns,
        index=[f"Chunk_{i+1}" for i in range(g_chunks)]
    )

    return results, chunk_means_df


results_moments, chunk_means = verify_first_moment_homogeneity(df_strain, g_chunks=5)

print("=== MANOVA First Moment Homogeneity Test ===")
for key, value in results_moments.items():
    print(f"{key}: {value}")

print("\n=== Chunk Means ===")
display(chunk_means)
```

### Numerical Interpretation

The dataset was divided into five consecutive non-overlapping chunks, each containing 1000 observations. The purpose of the test was to check whether the joint mean vector stayed constant across time.

Wilks' Lambda measures how much of the total scatter remains within the groups after accounting for the separation between group means. A value of \(\Lambda\) close to 1 suggests that the group means are similar. A smaller value suggests that the group means are separated.

Using Bartlett's chi-square approximation, the calculated test statistic is compared with a chi-square distribution with

\[
df=m(g-1)=3(5-1)=12.
\]

The computed p-value is much smaller than

\[
\alpha = 0.05.
\]

Therefore, the null hypothesis is rejected.

This means that the first moment is not homogeneous across the time blocks. The baseline center has shifted over time. In the context of structural health monitoring, this indicates that the baseline data cannot be treated as perfectly stationary in its joint mean.
