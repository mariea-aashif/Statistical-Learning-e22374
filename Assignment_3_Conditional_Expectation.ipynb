### 3)
By definition, $\mathbb{E}[X\mid Y]$ must be $\sigma(Y)$-measurable. This structural constraint means that the estimator can only depend on the macro-information revealed by $Y$, rather than the fine-grained individual outcomes of $\Omega$.

Since $\sigma(Y)$ cannot distinguish between outcomes within the partition cells

$$
\{\omega_1,\omega_2\}
\quad \text{and} \quad
\{\omega_3,\omega_4\},
$$

the random variable $\mathbb{E}[X\mid Y]$ must take a single, constant value on each of these respective sets.

Consequently, we express the conditional expectation as:

$$
\mathbb{E}[X\mid Y](\omega)=
\begin{cases}
a, & \omega\in\{\omega_1,\omega_2\},\\
b, & \omega\in\{\omega_3,\omega_4\},
\end{cases}
$$

where $a$ and $b$ are constants to be explicitly determined.

$$

Equating the two sides gives:

$$

### 1)
Let $\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$ follow a joint multivariate normal distribution with mean vector $\boldsymbol{\mu} = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix}$ and a symmetric, positive definite covariance matrix partitioned as:

$$
\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}
$$

The joint density function $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ for $\mathbf{z} \in \mathbb{R}^{n+m}$ is given by:

$$
f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right)
$$

---

### 2)
To find the marginal and conditional profiles, we decompose the joint quadratic form $Q = (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu})$ using the Schur complement of $\boldsymbol{\Sigma}_{YY}$ to perform a block matrix inversion. Let $\tilde{\mathbf{x}} = \mathbf{x} - \boldsymbol{\mu}_X$ and $\tilde{\mathbf{y}} = \mathbf{y} - \boldsymbol{\mu}_Y$. The inverse of the block covariance matrix can be factored as:

$$
\boldsymbol{\Sigma}^{-1} = \begin{bmatrix} \mathbf{I} & -\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1} \\ \mathbf{0} & \mathbf{I} \end{bmatrix}^T \begin{bmatrix} \mathbf{S}_{XX}^{-1} & \mathbf{0} \\ \mathbf{0} & \boldsymbol{\Sigma}_{YY}^{-1} \end{bmatrix} \begin{bmatrix} \mathbf{I} & -\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1} \\ \mathbf{0} & \mathbf{I} \end{bmatrix}
$$

where $\mathbf{S}_{XX} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$ is the Schur complement of $\boldsymbol{\Sigma}_{YY}$. Substituting this factorization into the quadratic form yields:

$$
Q = (\mathbf{y} - \boldsymbol{\mu}_Y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y) + (\mathbf{x} - \boldsymbol{\mu}^*)^T \mathbf{S}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*)
$$

where the conditional mean vector is defined as:

$$
\boldsymbol{\mu}^* = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y)
$$

---

### 3)
Using the definition of conditional density, we have:

$$
f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y})}{f_{\mathbf{Y}}(\mathbf{y})}
$$

The marginal density of $\mathbf{Y}$ captures the first term of our exponent decomposition:

$$
f_{\mathbf{Y}}(\mathbf{y}) = \frac{1}{\sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|}} \exp\left( -\frac{1}{2} (\mathbf{y} - \boldsymbol{\mu}_Y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y) \right)
$$

Dividing the joint density by this marginal density cancels out the $\mathbf{Y}$-marginal quadratic term from the exponential expression:

$$
Q_{\text{conditional}} = Q_{\text{joint}} - Q_{\text{marginal}} = (\mathbf{x} - \boldsymbol{\mu}^*)^T \mathbf{S}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*)
$$

For the coefficient factors, dividing the joint constant by the marginal constant gives:

$$
\frac{\frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}}}{\frac{1}{\sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|}}} = \frac{1}{\sqrt{(2\pi)^n \frac{|\boldsymbol{\Sigma}|}{|\boldsymbol{\Sigma}_{YY}|}}}
$$

Using the block determinant identity $|\boldsymbol{\Sigma}| = |\boldsymbol{\Sigma}_{YY}| \cdot |\mathbf{S}_{XX}|$, the fraction inside the root simplifies directly to $|\mathbf{S}_{XX}|$. Combining the simplified constant fraction and the remaining exponent profile yields the exact target conditional density function:

$$
f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \mathbf{S}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right)
$$

This matches the density function of a multivariate normal distribution:

$$
(\mathbf{X} \mid \mathbf{Y} = \mathbf{y}) \sim \mathcal{N}\left(\boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y), \; \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}\right)
$$

---

### 4)
By the Disintegration Theorem, the conditional expectation function $g(\mathbf{y}) = \mathbb{E}[\mathbf{X} \mid \mathbf{Y} = \mathbf{y}]$ is calculated as the first moment (mean) against the conditional probability density function:

$$
g(\mathbf{y}) = \int_{\mathbb{R}^n} \mathbf{x} f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) \, d\mathbf{x}
$$

Since $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$ is a valid Gaussian density centered at the shifting location parameter $\boldsymbol{\mu}^*$, evaluating this moment integral yields the mean parameter directly:

$$
g(\mathbf{y}) = \boldsymbol{\mu}(\mathbf{y}) = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y)
$$

Substituting the random vector $\mathbf{Y}$ back into our functional mapping gives the final system conditional expectation:

$$
\mathbb{E}[\mathbf{X} \mid \mathbf{Y}] = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{Y} - \boldsymbol{\mu}_Y)
$$

### 5)
The conditional covariance matrix corresponds to the variance profile of our conditional distribution, which is equal to the constant Schur block:

$$
\text{Var}(\mathbf{X} \mid \mathbf{Y} = \mathbf{y}) = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}
$$

Because this matrix is completely independent of the specific realized parameter value $\mathbf{y}$, taking its expected value across the outer state space leaves the terms completely unchanged:

$$
\mathbb{E}[\text{Var}(\mathbf{X} \mid \mathbf{Y})] = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}
$$

Applying the Law of Total Variance to check our structural alignment:

$$
\text{Var}(\mathbf{X}) = \mathbb{E}[\text{Var}(\mathbf{X} \mid \mathbf{Y})] + \text{Var}(\mathbb{E}[\mathbf{X} \mid \mathbf{Y}])
$$

Substituting our derived profiles into the right-hand side expressions confirms the breakdown:

$$
\text{Var}(\mathbb{E}[\mathbf{X} \mid \mathbf{Y}]) = \text{Var}\left(\boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{Y} - \boldsymbol{\mu}_Y)\right) = \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX} = \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}
$$

Summing the components together yields:

$$
\left(\boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}\right) + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX} = \boldsymbol{\Sigma}_{XX} = \text{Var}(\mathbf{X}) \quad \checkmark
$$

The terms cancel perfectly, showing that $\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$ acts as the **Explained Variance**, while the Schur complement matrix represents the **Unexplained Variance (Error)**.

### 6)
The differential entropy of an arbitrary $d$-dimensional multivariate normal random variable $\mathbf{V} \sim \mathcal{N}(\boldsymbol{\mu}_V, \boldsymbol{\Sigma}_V)$ is given by the standard formula:

$$
h(\mathbf{V}) = \frac{1}{2}\ln\left((2\pi e)^d |\boldsymbol{\Sigma}_V|\right) = \frac{d}{2}(1 + \ln(2\pi)) + \frac{1}{2}\ln|\boldsymbol{\Sigma}_V|
$$

Applying this formula to our system components:

**a)** The primary background uncertainty of the total concatenated random matrix $\mathbf{Z} \in \mathbb{R}^{n+m}$ is:
$$h(\mathbf{Z}) = \frac{1}{2}\ln\left((2\pi e)^{n+m} |\boldsymbol{\Sigma}|\right)$$

**b)** The residual system uncertainty profile of $\mathbf{X}$ after observing the context vector $\mathbf{Y}$ is governed by the conditional covariance matrix $\mathbf{S}_{XX}$:
$$h(\mathbf{X} \mid \mathbf{Y}) = \frac{1}{2}\ln\left((2\pi e)^n |\mathbf{S}_{XX}|\right) = \frac{1}{2}\ln\left((2\pi e)^n |\boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}|\right)$$

**c)** The mutual information transmission $I(\mathbf{X}; \mathbf{Y})$ measuring the dependency between the spaces is:
$$I(\mathbf{X}; \mathbf{Y}) = h(\mathbf{X}) - h(\mathbf{X} \mid \mathbf{Y})$$
$$I(\mathbf{X}; \mathbf{Y}) = \frac{1}{2}\ln\left((2\pi e)^n |\boldsymbol{\Sigma}_{XX}|\right) - \frac{1}{2}\ln\left((2\pi e)^n |\boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}|\right)$$
$$I(\mathbf{X}; \mathbf{Y}) = \frac{1}{2}\ln\left( \frac{|\boldsymbol{\Sigma}_{XX}|}{|\boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}|} \right)$$

```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# 1. Define foundational parameters for a 2D Gaussian system
mu = np.array([0.0, 0.0])
# Construct a positive definite block covariance matrix
sigma = np.array([[1.0, 0.7],
                  [0.7, 1.0]])

# Extract the individual scalar parameter blocks
sigma_xx = sigma[0, 0]
sigma_yy = sigma[1, 1]
sigma_xy = sigma[0, 1]
sigma_yx = sigma[1, 0]

# Generate synthetic data points from the joint space
n_samples = 5000
np.random.seed(42)
data = np.random.multivariate_normal(mu, sigma, n_samples)
x_samples, y_samples = data[:, 0], data[:, 1]

# 2. Compute theoretical conditional updates given a fixed observation profile
y_fixed = 1.2
mu_cond = mu[0] + (sigma_xy / sigma_yy) * (y_fixed - mu[1])
var_cond = sigma_xx - (sigma_xy**2 / sigma_yy)

# Filter samples where Y falls within a tight window near our target slice
condition_mask = np.abs(y_samples - y_fixed) < 0.1
x_conditional_samples = x_samples[condition_mask]

# 3. Compute Information-Theoretic Metrics
h_x = 0.5 * np.log(2 * np.pi * np.e * sigma_xx)
h_x_given_y = 0.5 * np.log(2 * np.pi * np.e * var_cond)
mutual_info = 0.5 * np.log(sigma_xx / var_cond)

print("="*60)
print("SYSTEM NUMERICAL VERIFICATION VERDICT")
print("="*60)
print(f"Conditional Distribution Target Center (mu*): {mu_cond:.4f}")
print(f"Conditional Distribution System Variance (S_xx): {var_cond:.4f}")
print("-"*60)
print(f"Unconditional System Entropy h(X)   : {h_x:.4f} nats")
print(f"Residual Conditional Entropy h(X|Y) : {h_x_given_y:.4f} nats")
print(f"Transferred Mutual Information I(X;Y): {mutual_info:.4f} nats")
print("="*60)

# 4. Generate Interactive Visualization Layout
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=("Joint Density Space P(X, Y)", "Marginal Profile P(Y)",
                    "Marginal Profile P(X)", f"Conditional Profile P(X | Y={y_fixed})")
)

# Top Left: Joint Scattering
fig.add_trace(go.Scatter(x=x_samples[::5], y=y_samples[::5], mode='markers',
                         marker=dict(color='blue', opacity=0.3, size=3), name="Joint Space"), row=1, col=1)
fig.add_shape(type="line", x0=-3, y0=y_fixed, x1=3, y1=y_fixed,
              line=dict(color="Red", width=2, dash="dash"), row=1, col=1)

# Top Right: Marginal Y Histogram
fig.add_trace(go.Histogram(y=y_samples, nbinsy=40, marker_color='green', name="Marginal Y"), row=1, col=2)

# Bottom Left: Marginal X Histogram
fig.add_trace(go.Histogram(x=x_samples, nbinsx=40, marker_color='orange', name="Marginal X"), row=2, col=1)

# Bottom Right: Shifting Conditional Distribution Profile Histogram
fig.add_trace(go.Histogram(x=x_conditional_samples, nbinsx=30, marker_color='red',
                           histnorm='probability density', name="Empirical X|Y"), row=2, col=2)

# Overlay theoretical Gaussian curve onto the empirical slice histogram
x_axis = np.linspace(-3, 3, 200)
y_gauss = (1.0 / np.sqrt(2 * np.pi * var_cond)) * np.exp(-0.5 * ((x_axis - mu_cond)**2) / var_cond)
fig.add_trace(go.Scatter(x=x_axis, y=y_gauss, mode='lines', line=dict(color='black', width=3), name="Theoretical PDF"), row=2, col=2)

fig.update_layout(title_text="Multivariate Normal Conditioning Analysis", height=700, width=900, showlegend=False)
fig.show()
```

### 1)
Since $X \sim \mathcal{N}(0, 1)$ and $Q \sim \mathcal{N}(0, 4)$ are independent normal random variables, any linear combination of $X$ and $Y = X + Q$ is also a linear combination of the independent normals $X$ and $Q$. Therefore, the vector $(X, Y)^T$ follows a joint multivariate normal distribution:

$$
\begin{bmatrix} X \\ Y \end{bmatrix} \sim \mathcal{N}(\boldsymbol{\mu}, \boldsymbol{\Sigma})
$$

To find the marginal distribution of $Y$, we look at its expectation and variance:
* **Mean:** $\mathbb{E}[Y] = \mathbb{E}[X + Q] = \mathbb{E}[X] + \mathbb{E}[Q] = 0 + 0 = 0$
* **Variance:** Since $X$ and $Q$ are independent, $\text{Var}(Y) = \text{Var}(X + Q) = \text{Var}(X) + \text{Var}(Q) = 1 + 4 = 5$

Thus, $Y \sim \mathcal{N}(0, 5)$.

To find the covariance $\text{Cov}(X, Y) = \Sigma_{XY}$:

$$
\Sigma_{XY} = \mathbb{E}[(X - \mathbb{E}[X])(Y - \mathbb{E}[Y])] = \mathbb{E}[X(X + Q)] = \mathbb{E}[X^2] + \mathbb{E}[XQ]
$$

Using independence ($\mathbb{E}[XQ] = \mathbb{E}[X]\mathbb{E}[Q] = 0$) and the definition of variance for a zero-mean variable ($\mathbb{E}[X^2] = \text{Var}(X) = 1$):

$$
\Sigma_{XY} = 1 + 0 = 1
$$

The complete joint parameters are:

$$
\boldsymbol{\mu} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}, \qquad \boldsymbol{\Sigma} = \begin{bmatrix} 1 & 1 \\ 1 & 5 \end{bmatrix}
$$

### 2)
A random variable $X$ is $\sigma(Y)$-measurable if and only if its value can be uniquely determined from a realization of $Y$. Mathematically, $\sigma(Y)$ contains information regarding the sum $X + Q$, which creates a pool of infinite combinations of $X$ and $Q$ that can yield the exact same value for $Y$.

In practical terms, if the power meter shows $Y = 5$, the engineer cannot isolate the true temperature because multiple physical scenarios map to this exact observation:
* **Overheating scenario:** The spindle could be dangerously hot ($X = 4$) with a standard mechanical cutting load ($Q = 1$).
* **Hard material scenario:** The spindle could be running at its optimal baseline temperature ($X = 0$) but cutting an exceptionally hard batch of material ($Q = 5$).

Because the mapping from $Y \to X$ is one-to-many, $X$ is not $\sigma(Y)$-measurable, creating fundamental informational ambiguity.

### 3)
Using the explicit projection property for joint multivariate normal distributions, the conditional expectation is:

$$
\mathbb{E}[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y)
$$

Substituting our known system parameters ($\mu_X = 0$, $\mu_Y = 0$, $\text{Cov}(X,Y) = 1$, $\text{Var}(Y) = 5$):

$$
\mathbb{E}[X \mid Y = y] = 0 + \frac{1}{5}(y - 0) = \frac{1}{5}y
$$

Expressing this as a functional random variable mapping:

$$
\mathbb{E}[X \mid Y] = \frac{1}{5}Y
$$

This expression reflects a "signal-to-total-variance" ratio. By rewriting the denominator using our structural blocks:

$$
\mathbb{E}[X \mid Y] = \left( \frac{\text{Var}(X)}{\text{Var}(X) + \text{Var}(Q)} \right) Y = \left( \frac{1}{1 + 4} \right) Y = \frac{1}{5}Y
$$

The weight assigned to our observation is directly proportional to how much of the total monitored power variance is driven by true thermal signal versus background mechanical noise.

### 4)
Let our optimal estimate be $\widehat{X} = \frac{1}{5}Y$. The estimation residual is defined as:

$$
Z = X - \widehat{X} = X - \frac{1}{5}Y
$$

We check the unbiased property by taking the expectation:

$$
\mathbb{E}[Z] = \mathbb{E}\left[X - \frac{1}{5}Y\right] = \mathbb{E}[X] - \frac{1}{5}\mathbb{E}[Y] = 0 - \frac{1}{5}(0) = 0
$$

To prove that the error is uncorrelated with our monitoring variable $Y$ (Orthogonality Principle), we compute their covariance:

$$
\text{Cov}(Z, Y) = \text{Cov}\left(X - \frac{1}{5}Y, \; Y\right) = \text{Cov}(X, Y) - \frac{1}{5}\text{Cov}(Y, Y)
$$

Substituting the variance of $Y$ ($\text{Cov}(Y,Y) = \text{Var}(Y) = 5$) and our joint covariance ($\text{Cov}(X,Y) = 1$):

$$
\text{Cov}(Z, Y) = 1 - \frac{1}{5}(5) = 1 - 1 = 0
$$

Since their covariance is zero, the residual error $Z$ is perfectly uncorrelated with the observation $Y$. Any informational patterns present in $Y$ have been completely extracted by the estimator.

### 7)
The manager's claim that $\widehat{X} = Y$ is a good approach overlooks the scaling of the noise term. If we evaluate the mean squared error of the manager's estimator, we find:

$$
\mathbb{E}[(Y - X)^2] = \mathbb{E}[Q^2] = \text{Var}(Q) = 4
$$

In contrast, our conditional expectation estimator yields an error equal to the conditional variance:

$$
\mathbb{E}[(\mathbb{E}[X\mid Y] - X)^2] = \sigma_{X|Y}^2 = 0.80
$$

The raw measurement $Y$ is an over-responsive estimator because it attributes every fluctuation in power directly to temperature changes.

The optimal minimum mean squared error (MMSE) estimator, $\mathbb{E}[X \mid Y] = \frac{1}{5}Y$, explicitly "shrinks" the measurement toward the global mean (zero). This shrinkage is a deliberate design feature dictated by the signal-to-noise ratio. Because $80\%$ of the variance in the power reading is driven by unobserved mechanical torque ($Q$), any large spike in $Y$ is statistically more likely to be caused by a heavy mechanical load than by a thermal spike. The estimator accounts for this by scaling back the measurement, protecting the system from generating false alarms due to normal torque variations.

```python
import numpy as np

# Define problem parameters
var_x = 1.0
var_q = 4.0
var_y = var_x + var_q  # 1 + 4 = 5
cov_xy = var_x         # 1.0

print("="*60)
print("CNC SPINDLE DIAGNOSTIC VERIFICATION")
print("="*60)

# 3) Verification of Signal-to-Noise Coefficient
coefficient = cov_xy / var_y
print(f"3) Optimal Estimator Equation: E[X | Y] = {coefficient:.2f} * Y")

# 4) Simulation-Based Proof of Orthogonality Principle
np.random.seed(42)
X = np.random.normal(0, np.sqrt(var_x), 100000)
Q = np.random.normal(0, np.sqrt(var_q), 100000)
Y = X + Q

X_hat = coefficient * Y
Z = X - X_hat

print(f"4) Simulation Results (100k samples):")
print(f"   Expected Error E[Z]      = {np.mean(Z):.4f} (Theoretical: 0)")
print(f"   Covariance Cov(Z, Y)     = {np.cov(Z, Y)[0, 1]:.4f} (Theoretical: 0)")

# 5 & 6) Information Theory Calculations
h_x = 0.5 * np.log(2 * np.pi * np.e * var_x)
var_cond = var_x - (cov_xy**2 / var_y)
h_x_cond_y = 0.5 * np.log(2 * np.pi * np.e * var_cond)
I_xy = h_x - h_x_cond_y
pct_reduction = (I_xy / h_x) * 100

print(f"5) Differential Entropies:")
print(f"   h(X)   = {h_x:.4f} nats")
print(f"   h(X|Y) = {h_x_cond_y:.4f} nats (Conditional Variance: {var_cond:.2f})")
print(f"6) Information Gain:")
print(f"   Mutual Information I(X;Y) = {I_xy:.4f} nats")
print(f"   Uncertainty Reduction     = {pct_reduction:.2f}%")

# High Noise Variance Case (Var(Q) = 100)
var_q_high = 100.0
var_y_high = var_x + var_q_high
var_cond_high = var_x - (cov_xy**2 / var_y_high)
h_x_cond_y_high = 0.5 * np.log(2 * np.pi * np.e * var_cond_high)
I_xy_high = h_x - h_x_cond_y_high
pct_reduction_high = (I_xy_high / h_x) * 100

print(f"   [With Var(Q)=100] New I(X;Y) = {I_xy_high:.4f} nats")
print(f"   [With Var(Q)=100] New Reduction = {pct_reduction_high:.2f}%")
print("="*60)
```
